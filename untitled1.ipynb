## to_datetime() e astype()

Convertendo a coluna `data` para o formato datetime64 do pandas. "[ns]" significa nano segundos.

```python
df_colheita['data'] = pd.to_datetime(df_colheita['data'], dayfirst=True)
```

```python
df_colheita.head()
```

```python
df_colheita.info()
```

Convertendo a coluna "data_extração" usando o método astype() para fazer conversão de tipo:

```python
df_colheita['data_colheita'] = df_colheita['data_colheita'].astype('datetime64[ns]')
```

```python
df_colheita.head()
```

```python
df_colheita.info()
```
